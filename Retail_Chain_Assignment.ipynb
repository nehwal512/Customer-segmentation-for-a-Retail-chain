```python
PRESENT = dt.datetime(2011,12,10)
new_data['InvoiceDate'] = pd.to_datetime(new_data['InvoiceDate'])
```

```python
new_data.head(5)
```

```python
rfm= new_data.groupby('CustomerID').agg({'InvoiceDate': lambda date: (PRESENT - date.max()).days,
                                        'InvoiceNo': lambda num: len(num),
                                        'Amount': lambda price: price.sum()})
```

```python
rfm.head(10)
```

```python
rfm.columns
```

```python
# Change the name of columns
rfm.columns=['frequency','recency','monetary']
```

```python
rfm['recency'] = rfm['recency'].astype(int)
```

```python
rfm.head(10)
```

```python
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
rfm['f_quartile'] = pd.qcut(rfm['frequency'], 4, ['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4','3','2','1'])
```

```python
rfm.head(10)
```

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
```

```python
rfm.head(10)
```

```python
# Filter out Top/Best cusotmers
rfm[rfm['RFM_Score']=='111'].sort_values('frequency', ascending=False).head(10)
```

```python
# Filter out Bottom/Bad cusotmers
rfm[rfm['RFM_Score']=='444'].sort_values('monetary', ascending=False).head(10)
```

```python
# loyal customers who buy often, but spend very little

rfm[rfm['r_quartile']=='4'].sort_values('recency', ascending=False).head(10)
```

## Prediction 

```python
import numpy as np
import matplotlib.pyplot as plt # for visualization
import seaborn as sns# for visualization
%matplotlib inline
```

```python
# Correlation Matrix and plot 
corrmat = new_data.corr() 
f, ax = plt.subplots(figsize =(9, 8)) 
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1) 
corrmat
```

```python
new_data.head(10)
```

```python
# indexes: X contain required Dependent Variable and Y contain Independent Variable 
X = new_data.iloc[:, [3, 4,]]
Y = new_data.iloc[:, 5]
```

```python
# Splitting the data 
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
# Linear Regression 
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# prediction
y_pred = regressor.predict(X_test)
```

```python
# model evaluation
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Mean square error 
rmse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)
```

```python
# printing values
print('Slope:' ,regressor.coef_)
print('Intercept:', regressor.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
```
